```python
# slice  assistant data
ass = []
ass.append(data.loc[range(18686,18973)])  # valve1
ass.append(data.loc[range(19069,19157)]) # valve2
ass.append(data.loc[range(19234,19316)]) # valve3
ass.append(data.loc[range(19418,19551)])# valve4
ass.append(data.loc[range(19633,19677)]) # valve5
ass.append(data.loc[range(19709,19842)]) # valve6
ass.append(data.loc[range(19975,20065)]) # valve7
ass.append(data.loc[range(20128,20224)]) # valve8
ass.append(data.loc[range(20311,20476)]) # valve9
ass.append(data.loc[range(20561,20680)]) # valve10
ass.append(data.loc[range(20789,21004)]) # valve11
ass.append(data.loc[range(21132,21209)]) # valve12
```

```python
### remove outliner 
fass = []
fnon = []

for k in range(0,12):
    tempAss = ass[k][(np.abs(stats.zscore(ass[k])) < 3).all(axis=1)]
    fass.append(tempAss)
    tempNon = non[k][(np.abs(stats.zscore(non[k])) < 3).all(axis=1)]
    fnon.append(tempNon)
```
