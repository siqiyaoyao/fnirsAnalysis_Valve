```python
non_result = currentNon.describe() 
#print(non_result)
ass_result = currentAss.describe() 
#print(ass_result)
```

```python
### show means
x = ass_result.columns[1:13] 
### ass
ass_y = ass_result.loc['mean'][1:13]
print(ass_y)
plt.scatter(x,ass_y,label ="Assistant")

###  non

non_y = non_result.loc['mean'][1:13]
# print(non_y)
plt.scatter(x,non_y,label ="Non")
plt.legend() 

fileName = str(currentIndex) +'.png'
plt.savefig(fileName)
plt.show() 
```
